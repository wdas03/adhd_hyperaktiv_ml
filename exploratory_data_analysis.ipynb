```python
import pandas as pd
import numpy as np
import sys
import os
```

```python
def get_patient_info():
    activity_dir = "hyperaktiv_with_controls"
    return pd.read_csv(activity_dir + "/patient_info.csv", delimiter=";")

def get_activity_data(patient_id):
    activity_dir = "hyperaktiv_with_controls/activity_data"
    fname = "{}/patient_activity_0{}.csv".format(activity_dir, patient_id) if patient_id < 10 else "{}/patient_activity_{}.csv".format(activity_dir, patient_id)

    print("Reading from {}:".format(fname))
    return pd.read_csv(fname, delimiter=';')
```
